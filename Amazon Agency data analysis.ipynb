### Understanding Your Variables

**1. Order Date:** The date on which customer placed the order.

**2. Agency Name :** Name of the agency involved in the transaction.

**3. Payment Date :** The date on which the payment for the order was confirmed.

**4. Payment Amount :** The total monetary amount paid by the customer for the order.

**5. Shipment Date :** The date on which product was dispatched from the warehouse.

**6. Product Category :**  Classification that groups products by type or industry for the reporting.

**7. ASIN :** Amazon Standard Identification Number- a unique product identifier used on Amazon.

**8. Title :** The product name displayed in the listing.

**9. UNSPSC :** United Nations Standard Products and Services Code - a global classification system for products and services.

**10. Brand Code :** A numeric or alphanumeric identifier used internally to track product brands.

**11. Brand :** The commercial name of the product manufacturer or trademark owner.

**12. Manufacturer :** The company or entity responsible for physically producing the item.

**13. Item Model Number :** Manufacturer assigned identifier for a special model of a product.

**14. Part number :** Identifier used to track individual parts.

**15. Product Condition :** Indicates if the product is new,used or renewed,etc.

**16. Purchase PPU :** Price per unit - the cost of single item before any discount.

**17. Item Quantity :** The total number of units purchased in the order line.

**18. Item Subtotal :** Cost calculation before taxes and shipping : purchase PPU * Item quantity.

**19. Item Shipping & Handling :** Fees charged for delivery,packing,and logistics services.

**20. Item Promotion :** Discount amount applied to the item from coupon codes,offers,etc.

**21. Item Tax :** Tax amount applied to the specific item based on local tax rules.

**22. Item Net Total :** Final payment amount for the item after adjusting : subtotal -promotion+shipping+tax.

**23. Seller Name :** The business or individual entity that listed and sold the product.
